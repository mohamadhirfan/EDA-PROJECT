### I want to know which Dealer sales which model...

```python
# crosstab : which is pandas it takes two categorical colums:
crosstab=pd.crosstab(cars_data['Model_name'],cars_data['Dealer_Name'][0:20])
crosstab # it will gives count only becuase it takes 2 categorical columns
```

```python
print(crosstab[0:5].plot(kind='bar',stacked=True))
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

## Observation:
>**From the above crosstab bar plot we observe that S-Class S 500 4MATIC model cars have most saled Dealer_name is James Motor Company .**
>**S-Class S 450 4MATIC model cars model saled by Price Pro**
